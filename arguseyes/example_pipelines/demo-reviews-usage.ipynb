```python
from arguseyes import ArgusEyes
from arguseyes.refinements import InputUsage
```

```python
eyes = ArgusEyes('demo-reviews', '../../mlruns')
```

This example operates on the [Amazon reviews pipeline](///edit/arguseyes/example_pipelines/amazon-reviews.py).

```python
with eyes.classification_pipeline_from_py_file('amazon-reviews.py') as pipeline:
    
    sources_with_usage = pipeline.compute(InputUsage())
```

```python
from arguseyes.templates.source import SourceType

reviews = [source for source in sources_with_usage if source.source_type==SourceType.FACTS][0].data
```

## Find all input records from the reviews table that were not used to train the classifier

```python
reviews[reviews['__arguseyes__is_used'] == False]
```

## Find all input records from the reviews table that were used to train the classifier

```python
reviews[reviews['__arguseyes__is_used'] == True]
```
